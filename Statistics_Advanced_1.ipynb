Question 4: What is a binomial distribution, and how is it used in probability?
-> Definition
A binomial distribution is a discrete probability distribution that describes the number of successes in a fixed number of independent Bernoulli trials, each with the same probability of success (p).

Conditions for a Binomial Distribution:
Fixed number of trials (n)

Two possible outcomes per trial → Success or Failure

Constant probability of success (p) for each trial

Independent trials




Formula
p(X=K) = (N/k)(1-P)n-k

n = number of trials

k = number of successes

p = probability of success

(nk) = combination formula




Example:
If you flip a fair coin 5 times (n = 5, p = 0.5),
the binomial distribution can give the probability of getting exactly 3 heads


Uses in Probability:
Modeling pass/fail tests (e.g., product quality checks)

Predicting number of successes in experiments

Business: Estimating conversion rates in marketing

Medicine: Predicting recovery rates in trials







Question 5: What is the standard normal distribution, and why is it important?

-> Definition:
The Standard Normal Distribution is a special case of the normal distribution with:

Mean (μ) = 0

Standard Deviation (σ) = 1

It is symmetric and bell-shaped.


Key Features:
The horizontal axis represents z-scores (standardized values):

z=x-μ/σ
The area under the curve = 1 (represents total probability).

About:

68% of values lie between z = -1 and z = 1

95% between z = -2 and z = 2

99.7% between z = -3 and z = 3


Why It’s Important:
Standardization – Converts different normal distributions into a common scale.

Probability Calculations – Z-tables are based on the standard normal distribution.

Hypothesis Testing – Many statistical tests (z-test, t-test) rely on it.

Real-World Applications – Used in quality control, finance, social sciences, etc.



Example:
If exam scores have μ = 70 and σ = 10,
a score of 85 has:

z = 85-70/10=1.5
→ This means it's 1.5 standard deviations above the mean.






Question 6: What is the Central Limit Theorem (CLT), and why is it critical in statistics?

->Definition:
The Central Limit Theorem (CLT) states that:

When we take many random samples from a population and compute their means, the distribution of these sample means will approximate a normal distribution, regardless of the population’s original distribution, provided the sample size is large enough (n ≥ 30 is a common rule).

Why It’s Critical:
Enables Hypothesis Testing – Many statistical tests assume normality.

Supports Confidence Intervals – We can estimate population parameters using sample data.

Practical for Real Data – Real-world populations are rarely perfectly normal.

 Key Points:
Works for any population distribution (normal or not).

The mean of the sampling distribution = population mean (μ).

The standard deviation of the sampling distribution = standard error:

Larger sample sizes → sampling distribution becomes more bell-shaped.

Example:
Population: Daily sales data (skewed distribution)

Take 50 random samples of size 40 each.

Plot sample means → they will form an approximately normal curve.




Question 7: What is the significance of confidence intervals in statistical analysis?
->Definition:
A confidence interval (CI) is a range of values, derived from sample data, that is likely to contain the true population parameter (e.g., mean, proportion) with a specified level of confidence (e.g., 95%).

Example:
If a 95% CI for average height is (165 cm, 172 cm), it means we are 95% confident that the true average height lies between 165 cm and 172 cm.

Significance:
Measures Precision of Estimates – A narrow CI means high precision; a wide CI means more uncertainty.

Gives a Range, Not Just a Point – Unlike a single sample mean, CI reflects the uncertainty due to sampling variability.

Supports Decision-Making – If a CI for a difference between two means includes 0, it may indicate no significant difference.

Interpretable in Real-World Terms – Helps non-statisticians understand uncertainty in estimates.

Example in Business:
A marketing team finds that the average sales increase after an ad campaign is 5% ± 2% (95% CI = 3% to 7%).

They can be fairly confident sales increase is positive.

Question 10: You are working as a data analyst for a retail company. The company has collected daily sales data for 2 years and wants you to identify the overall sales trend.

daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255, 235, 260, 245, 250, 225, 270, 265, 255, 250, 260]
Explain how you would apply the Central Limit Theorem to estimate the average sales with a 95% confidence interval. ● Write the Python code to compute the mean sales and its confidence interval. (Include your Python code and output in the code box below.)



-> Explanation
The Central Limit Theorem (CLT) says that if we take many random samples from a population, the distribution of the sample means will approach a normal distribution, even if the original data is not perfectly normal — as long as the sample size is large enough.

To estimate the average daily sales with a 95% confidence interval:

Mean sales → Average of the given sales data.

Standard deviation → Measures how much sales vary.

Standard error (SE) = Standard Deviation /
For a 95% confidence level, we use z-score = 1.96.

Confidence Interval (CI) = Mean±z×SE




```python
# Question 9




import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate 1000 random numbers from a normal distribution
mean = 50
std_dev = 5
data = np.random.normal(mean, std_dev, 1000)

# Step 2: Compute mean and standard deviation
calculated_mean = np.mean(data)
calculated_std_dev = np.std(data)

print("Calculated Mean:", calculated_mean)
print("Calculated Standard Deviation:", calculated_std_dev)

# Step 3: Draw a histogram
plt.hist(data, bins=30, edgecolor='black', alpha=0.7)
plt.title("Normal Distribution (mean=50, std=5)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()







# Question 10
import numpy as np

# Given daily sales data
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

# Convert to NumPy array
sales_array = np.array(daily_sales)

# Step 1: Calculate mean
mean_sales = np.mean(sales_array)

# Step 2: Calculate standard deviation
std_dev = np.std(sales_array, ddof=1)  # sample standard deviation

# Step 3: Calculate standard error
n = len(sales_array)
standard_error = std_dev / np.sqrt(n)

# Step 4: 95% confidence interval (z-score = 1.96)
z_score = 1.96
lower_bound = mean_sales - z_score * standard_error
upper_bound = mean_sales + z_score * standard_error

# Output
print(f"Mean Sales: {mean_sales:.2f}")
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```
